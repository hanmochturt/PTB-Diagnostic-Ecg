```python
# ============================================================================
# NO CAMBIES ESTE BLOQUE DE CÓDIGO
# ============================================================================
import warnings
warnings.filterwarnings('ignore')
import neurokit2 as nk
import matplotlib.pyplot as plt

# alternar ritmo cardiaco y nivel de ruido
ecg50 = nk.ecg_simulate(duration=10, noise=0.05, heart_rate=50)
plt.plot(ecg50)
plt.xlabel('muestras')
plt.ylabel("mV")
plt.title("50 latidos por minuto")
```
