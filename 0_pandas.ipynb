```python
import pandas as pd
import random
import numpy as np

def obtener_datos_ejemplo():
    # Base condition counts
    condition_counts = {"Myocardial infarction": 368, "Healthy control": 80, "n/a": 27,"Bundle branch block": 17, "Cardiomyopathy": 17,
        "Dysrhythmia": 16, "Hypertrophy": 7, "Valvular heart disease": 6, "Myocarditis": 4, "Stable angina": 2,  "Heart failure (NYHA 2)": 1,
        "Heart failure (NYHA 3)": 1, "Palpitation": 1,"Heart failure (NYHA 4)": 1,  "Unstable angina": 1
    }

    # Generate condition list
    condition_list = []
    for condition, count in condition_counts.items():
        condition_list.extend([condition] * count)

    # Patient ID
    patient_ids = [f"P{i+1:04d}" for i in range(len(condition_list))]

    # Sex (random)
    sex_list = random.choices(["Male", "Female"], k=len(condition_list))

    # Age ranges by condition (simplified)
    condition_age_ranges = {
        "Myocardial infarction": (45, 85),
        "Healthy control": (18, 60),
        "n/a": (18, 80),
        "Bundle branch block": (50, 80),
        "Cardiomyopathy": (30, 70),
        "Dysrhythmia": (40, 80),
        "Hypertrophy": (40, 70),
        "Valvular heart disease": (50, 85),
        "Myocarditis": (20, 50),
        "Stable angina": (50, 75),
        "Heart failure (NYHA 2)": (55, 80),
        "Heart failure (NYHA 3)": (55, 80),
        "Palpitation": (18, 40),
        "Heart failure (NYHA 4)": (55, 85),
        "Unstable angina": (50, 75)
    }

    # Assign age and BMI
    ages = []
    bmis = []

    for condition in condition_list:
        age_range = condition_age_ranges.get(condition, (18, 80))
        ages.append(random.randint(*age_range))
        bmis.append(round(np.random.normal(loc=26, scale=3), 1))  # Normal BMI distribution

    # Build DataFrame
    df_patients = pd.DataFrame({
        "Patient_ID": patient_ids,
        "Condition": condition_list,
        "Sex": sex_list,
        "Age": ages,
        "BMI": bmis
    })

    # Display sample
    return df_patients

df = obtener_datos_ejemplo()
```
